```python
import numpy as np
import seaborn as sns
from metropolis_hastings import MetropolisHastings
```

```python
target_pdf = lambda x: np.exp(-x ** 2 / 2) 
proposal_pdf = lambda x: np.random.normal(x,1)

state = 0
mh = MetropolisHastings(0, target_pdf, proposal_pdf, symmetric_poposal_pdf=True)

mh.sample_posterior(10**5)
 
sns.histplot(mh._samples)
mh.acceptance_rate()
```
